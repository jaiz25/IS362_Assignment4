## Windiest Weather


To find out which airport had the windiest weather on February 12, 2013, I will be using the weather.csv file.


```python
weather_file = pd.read_csv('weather.csv')
weather_file
```

```python
#sort out data that has the date February 12, 2013

weather_sort = weather_file[(weather_file['year'] == 2013) & (weather_file['month'] == 2) & (weather_file['day'] == 12)]
weather_sort.sort_values(by='wind_speed', ascending = False).head(5)
```

According to the sorted data, EWR airport was the windiest airport in February 12, 2013 with a wind speed of 1048.36058.  However, this number is an outlier.  Without the outlier, the windiest airport on that date is LGA.
